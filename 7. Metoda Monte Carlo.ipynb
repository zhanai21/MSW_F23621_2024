```python
import random
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.random.rand(N)
y = np.random.rand(N)

# Výpočet průměru a rozptylu x a y
print(np.mean(x), np.var(x))
print(np.mean(y), np.var(y))

# Vykreslení bodů (scatter plot) x a y
plt.plot(x, y, "o")  
plt.grid(True)
plt.show()  # Zobrazení grafu

# Vytvoření histogramu pro proměnnou x s 10 koši
num_bins = 10
n, bins, patches = plt.hist(x, num_bins)
plt.xlabel('x')
plt.ylabel('Pravděpodobnost')
plt.title('Histogram')
plt.show()
```
