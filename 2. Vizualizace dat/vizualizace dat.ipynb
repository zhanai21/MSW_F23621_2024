```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Načtení dat
df = pd.read_csv('heart.csv')

# Zobrazení prvních několika řádků dat
print(df.head(10))
```

```python

# Graf 1: Rozložení srdeční nemoci podle pohlaví
plt.figure(figsize=(6, 6))
sns.countplot(x='sex', hue='target', data=df)
plt.title('Rozložení srdeční nemoci podle pohlaví')
plt.xlabel('Pohlaví')
plt.ylabel('Počet')
plt.xticks(ticks=[0, 1], labels=['Ženský', 'Mužský'])
plt.legend(title='Srdeční nemoc', labels=['Chybí', 'Přítomná'])
plt.show()
# Tento graf ukazuje rozložení srdeční nemoci mezi muži a ženami.
```

```python

# Graf 2: Rozložení srdeční nemoci podle věku
plt.figure(figsize=(10, 6))
sns.histplot(df[df['target'] == 1]['age'], bins=20, color='skyblue', label='S nemocí')
sns.histplot(df[df['target'] == 0]['age'], bins=20, color='lightcoral', label='Bez nemoci')
plt.title('Rozložení srdeční nemoci podle věku')
plt.xlabel('Věk')
plt.ylabel('Frekvence')
plt.legend()
plt.show()
# Tento graf ukazuje rozložení věku pacientů s a bez srdeční nemoci.

```

```python

# Graf 3: Krabicový graf s očima krevního tlaku podle přítomnosti srdeční nemoci
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='trestbps', data=df)
plt.title('Krevní tlak podle přítomnosti srdeční nemoci')
plt.xlabel('Srdeční nemoc')
plt.ylabel('Krevní tlak')
plt.xticks(ticks=[0, 1], labels=['Chybí', 'Přítomná'])
plt.show()
# Tento graf ukazuje distribuci krevního tlaku mezi pacienty s a bez srdeční nemoci.


```

```python
# Graf 4: Histogram hladiny cukru v krvi podle přítomnosti srdeční nemoci
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='fbs', hue='target', multiple='dodge', bins=2, palette='Set2')
plt.title('Hladina cukru v krvi podle přítomnosti srdeční nemoci')
plt.xlabel('Hladina cukru (fbs > 120 mg/dl)')
plt.ylabel('Frekvence')
plt.legend(title='Srdeční nemoc', labels=['Chybí', 'Přítomná'])
plt.xticks(ticks=[0, 1], labels=['< 120 mg/dl', '> 120 mg/dl'])
plt.show()
```

```python

# Graf 5: Rozložení hladiny cholesterolu podle věku
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='chol', data=df, hue='target', palette='viridis')
plt.title('Rozložení hladiny cholesterolu podle věku')
plt.xlabel('Věk')
plt.ylabel('Hladina cholesterolu')
plt.legend(title='Srdeční nemoc', labels=['Chybí', 'Přítomná'])
plt.show()
```
